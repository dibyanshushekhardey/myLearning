```python
# example 1
from pylab import array

arr = array([77, 99, 44, 55, 22, 88, 11, 0, 66, 33])
nElems = 10

for j in arr:
    print(j)
print()

searchKey = 66

for j, val in enumerate(arr):
    if val == searchKey:
        break
if j == nElems:
    print("Can't find", searchKey)
else:
    print("found", searchKey)

searchKey = 55
print("Deleting", searchKey)
for j, val in enumerate(arr):
    if val == searchKey:
        break
k = j
while k < nElems - 1:
    arr[k] = arr[k+1]
    k += 1
nElems -= 1

j = 0
while j < nElems:
    print(arr[j])
    j += 1
```

```python
# Example 2
class LowArray:
    def __init__(self):
        self._v = []
    
    def setElem(self, index, value):
        if index < len(self._v):
            self._v[index] = value
        else:
            self._v.append(value)
    
    def getElem(self, index):
        return self._v[index]

arr = LowArray()
nElems = 0

arr.setElem(0, 77)
arr.setElem(1, 99)
arr.setElem(2, 44)
arr.setElem(3, 55)
arr.setElem(4, 22)
arr.setElem(5, 88)
arr.setElem(6, 11)
arr.setElem(7, 00)
arr.setElem(8, 66)
arr.setElem(9, 33)
nElems = 10

for j in range(nElems):
    print(arr.getElem(j))
print()

searchKey = 26
j = 0
while j < nElems:
    if arr.getElem(j) == searchKey:
        break
    j+= 1
if j == nElems:
    print("Can't find", searchKey)
else:
    print("Found", searchKey)

deleteKey = 55
print("Deleting element", deleteKey)

for j in range(nElems):
    if arr.getElem(j) == deleteKey:
        break

k = j
while k < nElems - 1:
    arr.setElem(k, arr.getElem(k+1))
    k += 1
nElems -= 1

for j in range(nElems):
    print(arr.getElem(j))
```

```python
# Example 3
class HighArray:
    def __init__(self):
        self._v =[]
    def find(self, searchKey):
        return searchKey in self._v
    def insert(self, value):
        self._v.append(value)
    def remove(self, value):
        try:
            self._v.remove(value)
            return True
        except ValueError:
            return False
    def display(self):
        print(self._v)

arr = HighArray()

#insert 10 items
arr.insert(77)	
arr.insert(99)
arr.insert(44)
arr.insert(55)
arr.insert(22)
arr.insert(88)
arr.insert(11)
arr.insert(0)
arr.insert(66)
arr.insert(33)

arr.display()

searchKey = 35
if arr.find(searchKey):
    print("Found", searchKey)
else:
    print("Can't find", searchKey)

print("Deleting 0, 55, and 99")
arr.remove(0)
arr.remove(55)
arr.remove(99)

arr.display()
```
